### Mean, median, maximum, ,minimum functions

### More pandas function

```python
def half(s):
    return s*0.5

iris[['Sales','Profit','Price']].apply(half)
```

```python
def double(s):
    return s*2

iris[['Sales','Profit','Price']].apply(double)
```
